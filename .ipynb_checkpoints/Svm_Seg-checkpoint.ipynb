```python
import numpy as np
from skimage import io
from sklearn.svm import SVC
```

```python
img = io.imread('Images/test.jpg')
```

```python
io.imshow(img)
io.show()
```

```python
rows, cols, bands = img.shape
classes = {'arbres': 0, 'vegetation': 1, 'sable': 2}
n_classes = len(classes)
palette = np.uint8([[255, 0, 0], [0, 255, 0], [0, 0, 255]])
```

```python
supervised = n_classes*np.ones(shape=(rows, cols), dtype=np.int)

supervised[180:200, 0:20] = classes['arbres']
supervised[0:20, 400:420] = classes['vegetation']
supervised[400:420, 380:400] = classes['sable']
```

```python
y = supervised.ravel()
train = np.flatnonzero(supervised < n_classes)
test = np.flatnonzero(supervised == n_classes)
```

```python
X = img.reshape(rows*cols, bands)
```

```python
clf = SVC(gamma='auto')
clf.fit(X[train], y[train])
y[test] = clf.predict(X[test])
supervised = y.reshape(rows, cols)

io.imshow(palette[supervised])
```
